```python
# from tabula import read_pdf
import tabula
import numpy as np
import re

#SSLpdf is in array 
SSLpdf = tabula.read_pdf('SSL Commercial Schedules.pdf', multiple_tables = True)

#get vessel name
SSLpdf_vessel_list = np.array(SSLpdf[0]).tolist()[0]
#to make the list same length as schedule
SSLpdf_vessel_list.insert(0, float('nan'))
SSLpdf_vessel_list.insert(0, float('nan'))
# print(SSLpdf_vessel_list)
#get vessel voyage
SSLpdf_voyage_list = np.array(SSLpdf[1]).tolist()[0]
#to make the list same length as schedule
SSLpdf_voyage_list.insert(0, float('nan'))
SSLpdf_voyage_list.insert(0, float('nan'))
# print(SSLpdf_voyage_list)
# print(SSLpdf_voyage_list)
# print(len(SSLpdf_voyage_list))
#get vessel schedule
SSLpdf_schedule_list = np.array(SSLpdf[2]).tolist()

#get number of rows
SSL_col_no = len(SSLpdf_vessel_list)
#get number of columns for schedule
SSL_row_no = len(SSLpdf[2])

SSL_schedule_list = []

# full vessel schedule in str
for x in range(0,SSL_row_no):
    SSL_schedule = SSLpdf_schedule_list[x]
#     print(SSL_schedule)
    SSL_schedule_str = [str(i) for i in SSL_schedule]
    SSL_schedule_list.append(SSL_schedule_str)
    
# print(SSL_schedule_list)

ssl_table_values = [] 

for i in range(2, SSL_col_no):
    vessel_name = SSLpdf_vessel_list[i]
    voyage_no = SSLpdf_voyage_list[i]
    
    for j in range(0, SSL_row_no):
        country = SSL_schedule_list[j][1]
        date = SSL_schedule_list[j][i]
        yes_bracket = bool(re.search("[()]", date))
        yes_singapore = bool(re.search("[SINGAPORE]", country))
        if yes_bracket == False:
            if yes_singapore == True:
                if date != 'nan':
                    ssl_table_values.append([vessel_name, voyage_no, country, date])
        
        
print(ssl_table_values)
```
